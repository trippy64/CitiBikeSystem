```python

# Add dependencies
import glob
import pandas as pd
```

```python
# use glob to get or return List[str]
# read the data from pd.read_csv(f) - to a dataframe  
# use the "for f in glob.glob()" to get a List[DataFrames]
# pd.concat() - grabs two or more and places in one pd.DataFrame()
bike_data = pd.concat([pd.read_csv(f) for f in glob.glob('rawdata/*.csv')], ignore_index = True)
```
